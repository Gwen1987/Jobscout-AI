```python
import matplotlib.pyplot as plt
import pandas as pd

# Summary data from all 3 simulations
data = {
    "Simulation": ["Sim 1", "Sim 2", "Sim 3"],
    "Val Accuracy": [97.71, 97.34, 95.47],
    "Val Precision": [95.05, 83.62, 51.96],
    "Val Recall": [55.49, 56.07, 84.39],
    "Val Loss": [0.0846, 0.0889, 0.1464]
}

df = pd.DataFrame(data)

# Plot Precision and Recall comparison
fig, ax = plt.subplots(figsize=(8, 5))
bar_width = 0.35
index = range(len(df))

# Bars
ax.bar(index, df["Val Precision"], bar_width, label="Precision", alpha=0.8)
ax.bar([i + bar_width for i in index], df["Val Recall"], bar_width, label="Recall", alpha=0.8)

# Axes & Labels
ax.set_xlabel("Simulation")
ax.set_ylabel("Score (%)")
ax.set_title("Validation Precision vs Recall Across Simulations")
ax.set_xticks([i + bar_width / 2 for i in index])
ax.set_xticklabels(df["Simulation"])
ax.legend()

# Save figure
plt.tight_layout()
plt.savefig("../images/precision_recall_comparison.png", dpi=300)
plt.show()
```
